```python
import pygam
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Clean_data.csv')
data
```

```python
glm_data = data.drop(['GroupID', 'Power', 'Brand', 'Gas', 'Region','Unnamed: 0', 'Density'], axis=1)
glm_data
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(glm_data, test_size=0.2)
```

```python
train.reset_index(inplace=True, drop=True)
train
```

```python
test.reset_index(inplace=True, drop=True)
```

```python
test
```

```python
from sklearn.model_selection import KFold
import numpy as np
import pandas as pd
from pygam import PoissonGAM, s,f

X = test.drop(['ClaimNb'], axis=1).values
y = test['ClaimNb'].values  

lambdas = [0.01, 0.1, 1, 10, 100]

kf = KFold(n_splits=5, shuffle=True, random_state=42)

results = {}

for lam in lambdas:
    fold_metrics = []
    for train_index, test_index in kf.split(X):
        X_train, X_val = X[train_index], X[test_index]
        y_train, y_val = y[train_index], y[test_index]
        
        gam = PoissonGAM(s(0) + s(1) + s(2) + s(7) + f(3) + f(4) + f(5) + f(6), lam = lam)
        gam.fit(X_train, y_train)
        
        y_pred = gam.predict(X_val)
        mse = np.mean((y_val-y_pred)**2)
        fold_metrics.append(mse) 
    

    results[lam] = np.mean(fold_metrics)

best_lambda = min(results, key=results.get)
print("Best lambda:", best_lambda)
print("Cross-validated MSE for each lambda:", results)
```
