```python
import pandas as pd 
import numpy as np 
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
import scipy.stats as sc
import matplotlib.pyplot as plt 
import seaborn as sns
import pygam
```

```python
data = pd.read_csv('Clean_data.csv')
data
```

```python
data['Power'].value_counts()
```

```python
data['Power_enc'].value_counts()
```

```python
glm_data = data.drop(['GroupID', 'Power', 'Brand', 'Gas', 'Region','Unnamed: 0'], axis=1)
glm_data
```

```python
glm_data['log_density'] = np.log(data['Density'])
```

```python
glm_data.drop(['Density'],axis=1, inplace=True)
```

```python
glm_data
```

```python
train, test = train_test_split(glm_data, test_size=0.2)
```

```python
train
```

```python
test
```

```python
Y = train['ClaimNb']
X = sm.add_constant(train.drop(['ClaimNb'], axis=1))
model = sm.GLM(Y, X, family=sm.families.Poisson())
results = model.fit()

print(results.summary())
```

```python
test1 = test.reset_index(drop=True)
```

```python
X_test = sm.add_constant(test1.drop(['ClaimNb'], axis=1))
predictions = results.predict(X_test)
print(predictions)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(test1['ClaimNb'], predictions)*59960/(59960-1-8)
```

```python
Y1 = train['ClaimNb']
X1 = sm.add_constant(train.drop(['ClaimNb', 'Brand_enc'], axis=1))
model1 = sm.GLM(Y1, X1, family=sm.families.Poisson())
results1 = model1.fit()

print(results1.summary())
```

```python
X_test1 = sm.add_constant(test1.drop(['ClaimNb','Brand_enc'], axis=1))
predictions1 = results1.predict(X_test1)
print(predictions1)
```
